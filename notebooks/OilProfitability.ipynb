# Descripción del proyecto

Para completar esta tarea, tendrás que realizar los siguientes pasos:

4. Elegir la región con el beneficio total más alto para los pozos petrolíferos seleccionados.


5. Tienes datos sobre muestras de crudo de tres regiones. Ya se conocen los parámetros de cada pozo petrolero de la región. 


6. Crea un modelo que ayude a elegir la región con el mayor margen de beneficio. Analiza los beneficios y riesgos potenciales utilizando la técnica bootstrapping.

# Condiciones:

- Solo se debe usar la regresión lineal para el entrenamiento del modelo.

- Después de la evaluación de riesgo, mantén solo las regiones con riesgo de pérdidas inferior al 2.5%. De las que se ajustan a los criterios, se debe seleccionar la región con el beneficio promedio más alto.

```python
# Proyecto de Selección de Pozos Petroleros - OilyGiant

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df_0 = pd.read_csv('/datasets/geo_data_0.csv')
df_1 = pd.read_csv('/datasets/geo_data_1.csv')
df_2 = pd.read_csv('/datasets/geo_data_2.csv')
```

```python
# ------------------ Entrenar y evaluar el modelo ------------------
def train_and_evaluate(df):
    features = df.drop(columns=['id', 'product'])
    target = df['product']
    X_train, X_valid, y_train, y_valid = train_test_split(features, target, test_size=0.25, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    rmse = np.sqrt(mean_squared_error(y_valid, predictions))
    return model, predictions, y_valid.reset_index(drop=True), rmse, predictions.mean()

model_0, preds_0, y_val_0, rmse_0, mean_0 = train_and_evaluate(df_0)
model_1, preds_1, y_val_1, rmse_1, mean_1 = train_and_evaluate(df_1)
model_2, preds_2, y_val_2, rmse_2, mean_2 = train_and_evaluate(df_2)
```

```python
# ------------------ Comparar con umbral de rentabilidad ------------------

budget = 100000000
price_per_unit = 4500
wells_selected = 200

# Un pozo debe generar al menos $500,000 -> 500000 / 4500 = 111.11 unidades
min_units_per_well = budget / (price_per_unit * wells_selected)
print(f"🔎 Un pozo debe producir al menos {min_units_per_well:.2f} mil barriles para evitar pérdidas.\n")

# Promedios predichos por región
promedios = {
    "Región 0": mean_0,
    "Región 1": mean_1,
    "Región 2": mean_2
}

# Comparación detallada
print("📊 Comparación con el mínimo necesario:")
print(f"{'Región':<10}{'Promedio predicho':>22}{'Diferencia vs mínimo':>25}{'¿Rentable?':>15}")
print("-" * 75)
for region, mean_pred in promedios.items():
    diff = mean_pred - min_units_per_well
    status = "✅ Sí" if diff >= 0 else "❌ No"
    print(f"{region:<10}{mean_pred:>22.2f}{diff:>25.2f}{status:>15}")
```

```python
# ------------------ Calcular ingreso y beneficio de los mejores 200 pozos ------------------

def calculate_profit(predictions, true_values):
    top_200_idx = predictions.argsort()[-200:]
    selected_true = true_values[top_200_idx]
    total_income = selected_true.sum() * price_per_unit
    net_profit = total_income - budget
    return total_income, net_profit, selected_true

# Calcular para cada región
income_0, profit_0, top_200_true_0 = calculate_profit(np.array(preds_0), np.array(y_val_0))
income_1, profit_1, top_200_true_1 = calculate_profit(np.array(preds_1), np.array(y_val_1))
income_2, profit_2, top_200_true_2 = calculate_profit(np.array(preds_2), np.array(y_val_2))

# Mostrar resultados
print("Resultados de los 200 pozos más prometedores por región:")
print(f"Región 0: Ingreso = ${income_0:,.2f}, Beneficio neto = ${profit_0:,.2f}")
print(f"Región 1: Ingreso = ${income_1:,.2f}, Beneficio neto = ${profit_1:,.2f}")
print(f"Región 2: Ingreso = ${income_2:,.2f}, Beneficio neto = ${profit_2:,.2f}")
```

```python
# ------------------ Bootstrapping para riesgo ------------------
def bootstrap_profit_correct(predictions, real_values, n_iterations=1000, sample_size=500, top_n=200, seed=42):
    np.random.seed(seed)
    profits = []

    data = pd.DataFrame({'prediction': predictions, 'real': real_values})

    for _ in range(n_iterations):
        # Tomar muestra aleatoria con reemplazo de 500 pozos
        sample = data.sample(n=sample_size, replace=True)

        # Seleccionar los 200 mejores por predicción
        top_200 = sample.nlargest(top_n, 'prediction')

        # Calcular ganancia con los valores reales
        profit = top_200['real'].sum() * price_per_unit - budget
        profits.append(profit)

    # Convertir a array y calcular métricas
    profits = np.array(profits)
    mean_profit = profits.mean()
    lower = np.percentile(profits, 2.5)
    upper = np.percentile(profits, 97.5)
    risk = (profits < 0).mean() * 100

    return profits, mean_profit, lower, upper, risk

boot_0, mean_p_0, low_0, high_0, risk_0 = bootstrap_profit_correct(preds_0, y_val_0)
boot_1, mean_p_1, low_1, high_1, risk_1 = bootstrap_profit_correct(preds_1, y_val_1)
boot_2, mean_p_2, low_2, high_2, risk_2 = bootstrap_profit_correct(preds_2, y_val_2)
```
